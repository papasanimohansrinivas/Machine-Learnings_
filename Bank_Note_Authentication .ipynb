# Wavelet Transformation

Images are stored  in format of (r,g,b) 3d matrices where r or g or b is a two dimensional matrix.

To do wavelet transformation of array [1,2,3,4] means (2,1),(4,3).

if a,b,c,d = [5,6,7,8]

then wavelet transformation means usually ((a+b)/2,(a+b)/2-b)),((c+d)/2,(c+d)/2-d)


```python
# example for wavelet transformation 

from __future__ import division

a,b,c,d = [141, 15, 100, 182]

m = (a+b)/2

(e,f) = (m,m-b)

print (e,f)

n = (c+d)/2

(g,h) = (n,n-d)

print (g,h)
```

# Terms

Image Skewness = brightness of image in laymans terms 


Image kurtosis = greyness of image or resolution


more greyness > 0 

less greyness < 0 


Image variance = difference between adjacent two pixel values and useful for edge detection


Image entropy =  tells us whether it is easy to compress the image or not

# Loading data set ......

```python

variances = []
entropy = []
skewedness = []
curtosis = []
classes =[]

for y in p:
    a,b,c,d,e = map(float,y.split(","))
    variances.append(a)
    entropy.append(d)
    skewedness.append(b)
    curtosis.append(c)
    classes.append(int(e))
    
import seaborn
%matplotlib inline
```

Below graph indecates there are almost equal number of genuine and forged bank note samples from data set


```python
seaborn.distplot(classes)
```
