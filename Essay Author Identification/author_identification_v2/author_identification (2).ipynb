```
def match(y,t4):
  
  assert(len(y)==len(t4))
  
  c = 0
  
  h = t4.values.tolist()
  
  h2 = list(y)
  
  def comp(list_1,list_2):
    for n in range(len(list_1)):
      if list_1[n]!=list_2[n]:
        return False
    return True
    
  for u in range(len(h)):
    if comp(h[u],h2[u]):
      c+=1
  
  print((c+0.0)/len(t4))
  
```
